```python
import pandas as pd
import numpy as np
df = pd.read_csv('1900_2021_disasters.csv')
```

```python
df.head()
```

```python
def clean_data(df):
    # Drop unnecessary columns
    df = df.drop(columns=['Glide', 'Aid Contribution', 'OFDA Response', 'Appeal', 'Declaration', 'Total Deaths', 'No Injured', 'No Affected', 'No Homeless', 'Total Affected', "Insured Damages ('000 US$)", "Total Damages ('000 US$)", 'Adm Level', 'Admin1 Code', 'Admin2 Code', 'Geo Locations'])
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    # Drop rows where 'Disaster Subgroup' is 'Biological'
    df = df[df["Disaster Subgroup"].str.contains("Biological") == False]
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```
